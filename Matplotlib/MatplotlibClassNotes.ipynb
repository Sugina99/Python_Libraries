```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[20,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title("Weather Info")
plt.plot(days,max_t)
plt.plot(days,min_t,color='yellow')
plt.plot(days,avg_t)
```

```python
plt.plot(days, max_t,label="max")
plt.plot(days, min_t,label="min")
plt.plot(days, avg_t,label="average")
plt.legend(loc="best",shadow='True',fontsize='small') #legend specification
plt.grid() #to identify the specific values by creating grid
plt.show()
```

```python
# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
loss=[78,45,32,10]
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.xlabel("major companies")
plt.ylabel("Revenue")
plt.title("Revenue model for Top companies")
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')
plt.legend()
plt.show()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xlabel("major companies")
plt.ylabel("Revenue")
plt.title("Revenue model for Top companies")
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)
plt.xlabel("major companies")
plt.ylabel("Revenue")
plt.title("Revenue model for Top companies")
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)
plt.xlabel("major companies")
plt.ylabel("Revenue")
plt.title("Revenue model for Top companies")
plt.bar(xpos,revenue,width=0.2,label='Revenue')
plt.bar(xpos+0.4,profit,width=0.2,label='Profit')
plt.bar(xpos+0.2,loss,width=0.2,label='loss')
plt.legend()
plt.show()
```
